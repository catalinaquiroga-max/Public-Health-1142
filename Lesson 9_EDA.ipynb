# Lesson 9: Exploratory Data Analysis

```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/juanklopper/TutorialData/main/heart_failure.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

## Tasks

### Task 1: Anaemia Analysis 

```python
df['anaemia'].unique()
```

```python
df['anaemia'].value_counts()
```

```python
df['anaemia'].value_counts(normalize=True)
```

### Task 2: Diabetes Analysis 

```python
df['diabetes'].unique()
```

```python
df['diabetes'].value_counts()
```

```python
df['diabetes'].value_counts(normalize=True)
```

### Task 3: Anemia vs Diabetes 

```python
df.groupby('diabetes').anaemia.value_counts()
```

```python
df.groupby('diabetes').anaemia.value_counts(normalize=True) * 100
```

### Task 4: Crosstab (Diabetes Vs Anaemia)

```python
pd.crosstab(df.diabetes, df.anaemia)
```

```python
pd.crosstab(df.diabetes, df.anaemia, normalize=True) * 100
```

### Task 5: Anaemia Vs Death

```python
pd.crosstab(df.death, df.anaemia, normalize=True) * 100
```

### Task 6: Mean Age

```python
df.age.mean()
```

```python
df.groupby('death').age.mean()
```

### Task 7: Median Ejection Fraction 

```python
df.groupby('anaemia')['ejection_fraction'].median()
```

### Task 8: Dispersion Measures 

```python
df.age.min()
```

```python
df.age.max()
```

```python
df.age.max() - df.age.min()
```

```python
df.age.var(ddof=1)
```

```python
df.age.std(ddof=1)
```

```python
stats.iqr(df.age)
```

### Task 9: Variance by Anaemia 

```python
df.groupby('anaemia').age.var(ddof=1)
```

### Task 10: Correlation

```python
df.age.corr(df['ejection_fraction'])
```

### Task 11: Describe Age

```python
df.age.describe()
```
