# Lesson 10: Data Visualization (Seaborn)

```python
import pandas as pd
import seaborn
from matplotlib import pyplot
```

```python
seaborn.set_style('whitegrid')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/juanklopper/TutorialData/main/heart_failure.csv')

df.head()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/juanklopper/TutorialData/main/heart_failure.csv')

df.head()
```

## Relational Plot: Age vs Platelets

```python
seaborn.scatterplot(
    data=df,
    x='age',
    y='platelets'
).set(
    title='Age vs Platelets',
    xlabel='Age',
    ylabel='Platelets'
)
```

## Task 1: Platelets vs Serum Sodium (by Diabetes)

```python
seaborn.scatterplot(
    data=df,
    x='platelets',
    y='serum_sodium',
    hue='diabetes'
).set(
    title='Platelets vs Serum Sodium',
    xlabel='Platelets',
    ylabel='Serum Sodium'
)
```

## Task 2: Platelets Distribution Histogram 

```python
# Minimum and maximum values
df['platelets'].min()
df['platelets'].max()

# Histogram
seaborn.displot(
    data=df,
    y='platelets'
)

pyplot.title('Platelet Distribution')
pyplot.xlabel('Frequency')
pyplot.ylabel('Platelets')
```

## Distrubution Plot: Platelets

```python
seaborn.displot(
    df,
    y='platelets',
    bins=range(0, 850000, 50000)
).set(
    title='Platelet Distribution',
    ylabel='Platelets',
    xlabel='Frequency'
)
```

## Categorical Plot: Diabetes Count 

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    kind='count'
).set(
    title='Diabetes Count',
    xlabel='Diabetes',
    ylabel='Frequency'
)
```

## Task 3: Diabetes Vs Death

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    hue='death',
    kind='count'
)
```

## Box Plot: Age Vs Diabetes

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    kind='box'
).set(
    title='Age vs Diabetes',
    xlabel='Diabetes',
    ylabel='Age'
)
```

## Violin Pilot: Age Vs Diabetes by Death

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    hue='death',
    kind='violin'
).set(
    title='Age vs Diabetes by Death',
    xlabel='Diabetes',
    ylabel='Age'
)
```
