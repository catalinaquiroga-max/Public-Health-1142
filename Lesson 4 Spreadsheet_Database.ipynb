# QUIZ QUESTIONS

1. What is the difference between a spreadsheet and a database?

Spreadsheet: used for small datasets and simple calculations in a flat structure. 

Database: is designed to manage large datasets using many related tables and complex systems.

2. What are the advantages and disadvantages of using a spreadsheet for health data management?

Advantages:

- Easy to learn and use

- Quick data entry and calculations

- Suitable for small datasets

- Built-in visualization tools

Disadvantages:

- Limited scalability

- Weak data integrity controls

- Restricted multi-user functionality

- Data inconsistency in bigger datasets

3. What are the advantages and disadvantages of using a database for health data management?

Advantages:

- Can handle large volumes of data

- Maintains data integrity

- Supports multiple users

- Allows complex querying and reporting

Disadvantages:

- More complex to design and maintain

- Requires technical knowledge
