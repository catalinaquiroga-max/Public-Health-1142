## Task 3 Set Difference Comparison


````python
set_A = {'a', 'b', 'c', 'd', 'e', 'f'}
set_B = {'a', 'c', 'f', 'g'}

# A difference B
difference_A_B = set_A.difference(set_B)

# B difference A
difference_B_A = set_B.difference(set_A)

difference_A_B, difference_B_A
````

-Set Differences are not the same


-Set Differences not symmetric

## Quiz Questions and Answers





### 1. How do you define a list in Python?

A list is defined using square brackets.

```python
my_list = [1, 2, 3]
```

---

### 2. How do you access the third element of `my_list`?

Python uses zero-based indexing.

```python

```

### 3. How do you add `"new_item"` to the end of `my_list`?

