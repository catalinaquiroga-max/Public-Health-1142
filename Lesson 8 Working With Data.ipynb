## Task 10: Modify Age and Create AgeHR 

```python
df.Age = df.Age + 2
df.Age.head()
```

```python
def add2(x):
    return x + 2
```

```python
df.Age = df.Age.apply(add2)
df.Age.head()
```

```python
df['AgeHR'] = df.Age * df.HR
df[['Age', 'HR', 'AgeHR']].head()
```
