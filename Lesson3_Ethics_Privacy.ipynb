## Quiz Questions

1. What is the primary goal of data anonymization?

The primary goal of data anonymization is to protect individual privacy while still allowing data to be used for research and analysis.

2. What is the Havasupai Tribe case about?

It involved researchers using blood samples for purposes beyond what the tribe originally consented to, violating informed consent and cultural trust.

3. What was the largest health data breach in history?

The Anthem data breach in 2015, which affected approximately 78.8 million individuals.

4. What is the pulse oximeter controversy about?

It revealed racial bias in medical devices, as pulse oximeters were less accurate in detecting low oxygen levels in Black patients.

5. What is the role of an Institutional Review Board (IRB)?

An IRB reviews and approves research involving human subjects to ensure ethical standards and participant protection.

6. What is the difference between spreadsheet software and database software?

Spreadsheet software manages smaller datasets in table form, while database software handles large, structured datasets with greater security and organization.

7. What is a conflict of interest in public health research?

A conflict of interest occurs when a secondary interest, such as financial gain, may influence research decisions.

8. What is informed consent?

Informed consent means participants are given clear information about a study and voluntarily agree to participate.

9. What is data masking?

Data masking involves replacing or removing identifiable information to protect privacy.

10. What is k-anonymity?

K-anonymity ensures that each individual record cannot be distinguished from at least k other records in a dataset.

11. What is the main concern about conflicts of interest?

They can introduce bias into research and reduce public trust.

12. What is the first step in managing conflicts of interest?

The first step is disclosure and transparency.

13. What is the role of independent oversight?

Independent oversight monitors research to ensure ethical conduct and reduce bias.

14. What is one strategy for mitigating conflicts of interest?

Public disclosure of any potential conflicts.

15. What is the potential impact of not properly managing conflicts of interest?

It can lead to biased research findings, loss of credibility, and reduced public trust.
