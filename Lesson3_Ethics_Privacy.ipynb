## Task 2 Breach in Health Data Security 

One of the biggest breaches in health data secutiry happened in 2017, during the WannaCry ransome attack. This affected the United Kingdom National Health Service, where more than 300,000 computer across hundreds of countries were infected and thousand of NHS appointments and surgeries were canceled. 

The hackers made use of old software systems that were yet to be updated had many security gaps, and as a result patients information became impossible to see so many services were disrupted. 

This clearly delayed treatments, financial looses and increased risk to patient safety. 

If there were more regular system updates and a stronger cybersecurity this could have been prevented by the NHS. 

## TASK 3 Bias in Health Data 

An example of bias in health date in the US involved a healthcare procedure used to identify the patients that needed more medical care. This procedure used healthcare spedning as a way to calculate the severity of an illness.

Since the majority of patients that have lower healthcare spending due to unequal access to care are Black patients, the algorithm directly assumed that these patients were healthier than White patients with similar medical cases. 

In effect, less Black patients were helped and offered additional treatment and support. 

This could have been prevented by testing the procedure in many different populations, so that this insures equal representation and date collection in further model evolution.

## TASK 4 ANONYMIZATION OF HEALTH DATA 

An example of anonymization in health data is the practice k-anonymity in research datasets. With this way of researching the date is grouped in a way that every record is alike to the other K record. 

For instance, instead of writing down a patient exact age (37), the data may just show an age range. This can protect the patients actual identity while at the same time allowing the researching to understand patterns within the data.
